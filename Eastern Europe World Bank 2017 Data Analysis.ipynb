```python
#Birth Rate vs Death Rate vs Populations Density (Graph Only)

# Setting the positions and width for the bars

graphData=teamDataCleaned_df.loc[['ALB','BIH','HRV','GRC','HUN','LVA','LTU','SRB'],['CountryName','NetBirthDeath','MigrationRate','PopulationGrowth']]

pos = list(range(len(graphData['NetBirthDeath']))) 
width = 0.25

# Plotting the bars
fig, ax = plt.subplots(figsize=(10,6))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        graphData['NetBirthDeath'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='red', 
        # with label the first value in first_name
        label=graphData['CountryName'][0],
        align='edge')

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['post_score'] data,
        graphData['PopulationGrowth'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='seagreen', 
        # with label the third value in first_name
        label=graphData['PopulationGrowth'][2],
        align='edge')

# Set the y axis label
#ax.set_ylabel('Numbers in Millions')

# Set the chart's title
ax.set_title('Net Birth/Death Rate vs PopulationGrowth')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(graphData['CountryName'], rotation=45)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([-1.5, 1])

# Adding the legend and showing the plot
plt.legend(['Net Birth/Death Rate', 'Population Growth'], loc='upper left')
plt.grid(True)
plt.show()
```
